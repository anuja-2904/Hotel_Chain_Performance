```python
df.isna().sum()
```

```python
df['Hotel_Type'] = df['Hotel_Type'].fillna("Unknown")
```

```python
df['Avg_Room_Price'] = df['Avg_Room_Price'].fillna(df['Avg_Room_Price'].mean())
```

```python
df['Customer_Rating'] = df['Customer_Rating'].fillna(df['Customer_Rating'].mean())
```

```python
df['Cancellations'] = df['Cancellations'].fillna(df['Cancellations'].mean())
```

```python
df['City'] = df['City'].replace('mumbai','Mumbai')
df['City'] = df['City'].replace('pune','Pune')
df['City'] = df['City'].replace('delhi','Delhi')
df['City'] = df['City'].replace('hyderabad','Hyderabad')
df['City'] = df['City'].replace('chennai','Chennai')
df['City'] = df['City'].replace('bengaluru','Bengaluru')

df['Event_Season'] = df['Event_Season'].replace('off season','Off Season')
df['Event_Season'] = df['Event_Season'].replace('wedding','Wedding')
df['Event_Season'] = df['Event_Season'].replace('festive','Festive')
df['Event_Season'] = df['Event_Season'].replace('conference','Conference')
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.shape
```

```python
df['Date'] = pd.to_datetime(df["Date"])
```

```python
df['Month'] = df['Date'].dt.month
```

```python
df['Year'] = df['Date'].dt.year
```

```python
df["month_name"]=df["Date"].dt.month_name()
```

```python
df["day_name"]=df["Date"].dt.day_name()
```

```python
df["quarter"]=df["Date"].dt.quarter
```
